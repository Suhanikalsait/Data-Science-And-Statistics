# Creating a Data Frame

```python
df = pd.DataFrame([[27, 15, 10, 5],[2, 10, 18, 22],[15, 26, 0, 20],[7, 11, 13, 20]],
                        columns = ["CD","DSS","DBMS","CAO"]
                        )
```

```python
df
```

```python
df.shape
```

```python
df.size
```

```python
df.ndim
```

```python
df2 = pd.DataFrame([[27, 10, 11, 0]],
                        columns = ["CD","DSS","DBMS","CAO"]
                        )
```

```python
df3 = pd.concat([df, df2], ignore_index=True)
```

```python
df3
```

# Adding Attribute (Colums) to Data Frame

```python
df3["DM"]=[12, 18, 2, 8, 11]
```

```python
df3
```

# Deleting Record (Row) From Data Frame

```python
df4= df3.drop(index=[1])
```

```python
df4
```

```python
print("Mean of DSS:", df4["DSS"].mean())
```

```python
print("Mean of DSS:", df4["DSS"].median())
```

```python
print("Mean of DSS:", df4["DSS"].mode())
```
